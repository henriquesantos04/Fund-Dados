```python
import pandas as pd
import matplotlib.pyplot as pit# type: ignore
```

```python
questionario = pd.read_csv("C:/Users/20221enpro0294/Documents/questionario.csv", sep = ";", decimal = ",")  
```

```python
tabela_sexo = questionario["Sexo"].value_counts()
tabela_sexo
```

```python
tabela_sexo = questionario["Sexo"].value_counts()/len(questionario)*100
tabela_sexo
```

```python
pit.bar(tabela_sexo.index, tabela_sexo.values)
pit.title("Distribuição de Sexo")
pit.xlabel("Sexo")
pit.ylabel("Quantidade de Pessoas")
```

```python
pit.barh(tabela_sexo.index, tabela_sexo.values)
pit.title("Distribuição de Sexo")
pit.ylabel("Sexo")
pit.xlabel("Quantidade de Pessoas")
```

```python
pit.hist(questionario["Alt"], bins = 8)
pit.title("Distribuição das Alturas")
pit.ylabel("Frequência")
pit.xlabel("Altura")
```
